```python
import cv2
import numpy as np
import os
import random

# --- Helper function for varied colors ---
def get_varied_color(base_color, variation_range=20):
    """
    Generates a new RGB color by perturbing the base_color components within a specified range.
    """
    varied_color = []
    for component in base_color:
        new_component = component + random.randint(-variation_range, variation_range)
        varied_color.append(max(0, min(255, new_component))) # Clamp between 0 and 255
    return tuple(varied_color)

# --- Realistic Color Palettes ---
skin_background_colors = [
    (255, 224, 189), (240, 207, 166), (222, 184, 135), (205, 133, 63),
    (250, 235, 215), (255, 239, 213), (253, 245, 230), (255, 250, 205),
    (255, 248, 220), (245, 222, 179)
]

lesion_colors = [
    (139, 0, 0), (178, 34, 34), (220, 20, 60), (255, 0, 0), (255, 69, 0),
    (101, 67, 33), (128, 0, 0), (88, 41, 0), (50, 25, 0),
    (0, 0, 0), (25, 25, 25), (50, 50, 50), (75, 75, 75),
    (128, 0, 128), (147, 112, 219), (75, 0, 130)
]

# --- Dataset Generation Loop ---
# Create folders
os.makedirs("dataset/images/train", exist_ok=True)
os.makedirs("dataset/labels/train", exist_ok=True)

img_size = 640
num_images = 100

for i in range(num_images):

    # Randomly select a skin background color
    background_color = random.choice(skin_background_colors)
    img = np.full((img_size, img_size, 3), background_color, dtype=np.uint8)

    num_objects = random.randint(1, 4)
    label_lines = []

    for _ in range(num_objects):
        shape_type = random.choice(['rectangle', 'circle', 'ellipse', 'polygon'])

        # Randomly select a lesion color
        base_lesion_color = random.choice(lesion_colors)

        if shape_type == 'rectangle':
            w, h = random.randint(40, 150), random.randint(40, 150)
            x = random.randint(0, img_size - w)
            y = random.randint(0, img_size - h)

            cv2.rectangle(img, (x, y), (x + w, y + h), base_lesion_color, -1)

            # Add concentric rectangles for texture
            if w > 10 and h > 10:
                for _ in range(random.randint(1, 2)): # One or two concentric shapes
                    reduce_w, reduce_h = random.randint(5, 15), random.randint(5, 15)
                    inner_w, inner_h = w - reduce_w, h - reduce_h
                    if inner_w > 0 and inner_h > 0: # Ensure dimensions are positive
                        inner_x = x + reduce_w // 2
                        inner_y = y + reduce_h // 2
                        cv2.rectangle(img, (inner_x, inner_y), (inner_x + inner_w, inner_y + inner_h), get_varied_color(base_lesion_color), -1)

            x_center_yolo = (x + w / 2) / img_size
            y_center_yolo = (y + h / 2) / img_size
            width_yolo = w / img_size
            height_yolo = h / img_size

        elif shape_type == 'circle':
            r = random.randint(20, 75) # Radius
            cx = random.randint(r, img_size - r)
            cy = random.randint(r, img_size - r)

            cv2.circle(img, (cx, cy), r, base_lesion_color, -1)

            # Add concentric circles for texture
            if r > 5:
                for _ in range(random.randint(1, 2)): # One or two concentric shapes
                    inner_r = r - random.randint(3, 10)
                    if inner_r > 0: # Ensure radius is positive
                        cv2.circle(img, (cx, cy), inner_r, get_varied_color(base_lesion_color), -1)

            x_center_yolo = cx / img_size
            y_center_yolo = cy / img_size
            width_yolo = (2 * r) / img_size
            height_yolo = (2 * r) / img_size

        elif shape_type == 'ellipse':
            major_axis = random.randint(30, 100)
            minor_axis = random.randint(15, min(major_axis - 1, 90))
            cx = random.randint(major_axis, img_size - major_axis)
            cy = random.randint(major_axis, img_size - major_axis)
            angle = random.randint(0, 180) # Angle in degrees

            cv2.ellipse(img, (cx, cy), (major_axis, minor_axis), angle, 0, 360, base_lesion_color, -1)

            # Add concentric ellipses for texture
            if major_axis > 5 and minor_axis > 5:
                for _ in range(random.randint(1, 2)): # One or two concentric shapes
                    inner_major_axis = major_axis - random.randint(3, 10)
                    inner_minor_axis = minor_axis - random.randint(3, 10)
                    if inner_major_axis > 0 and inner_minor_axis > 0: # Ensure dimensions are positive
                        cv2.ellipse(img, (cx, cy), (inner_major_axis, inner_minor_axis), angle, 0, 360, get_varied_color(base_lesion_color), -1)

            x_center_yolo = cx / img_size
            y_center_yolo = cy / img_size
            width_yolo = (2 * max(major_axis, minor_axis)) / img_size
            height_yolo = (2 * max(major_axis, minor_axis)) / img_size

        elif shape_type == 'polygon':
            num_vertices = random.randint(3, 7)
            points = []
            # Define a rough bounding box for the polygon vertices
            poly_cx = random.randint(50, img_size - 50)
            poly_cy = random.randint(50, img_size - 50)
            poly_size = random.randint(30, 80) # Max deviation from center

            min_x, min_y = img_size, img_size
            max_x, max_y = 0, 0

            for _ in range(num_vertices):
                px = random.randint(max(0, poly_cx - poly_size), min(img_size - 1, poly_cx + poly_size))
                py = random.randint(max(0, poly_cy - poly_size), min(img_size - 1, poly_cy + poly_size))
                points.append([px, py])
                min_x = min(min_x, px)
                min_y = min(min_y, py)
                max_x = max(max_x, px)
                max_y = max(max_y, py)

            points = np.array(points, np.int32)
            points = points.reshape((-1, 1, 2)) # Reshape for cv2.fillPoly

            cv2.fillPoly(img, [points], base_lesion_color)

            # Add noise for polygon texture
            mask = np.zeros(img.shape[:2], dtype=np.uint8)
            cv2.fillPoly(mask, [points], 255)

            noise_layer = np.random.randint(-20, 21, img.shape, dtype=np.int16) # Random integers between -20 and 20
            textured_img = img.astype(np.int16) + noise_layer
            textured_img = np.clip(textured_img, 0, 255).astype(np.uint8)

            # Apply texture only where the mask is 255 (i.e., within the polygon)
            img[mask == 255] = textured_img[mask == 255]

            x_center_yolo = ((min_x + max_x) / 2) / img_size
            y_center_yolo = ((min_y + max_y) / 2) / img_size
            width_yolo = (max_x - min_x) / img_size
            height_yolo = (max_y - min_y) / img_size

        label_lines.append(f"0 {x_center_yolo} {y_center_yolo} {width_yolo} {height_yolo}")

    # Save image
    img_name = f"img_{i}.jpg"
    cv2.imwrite(f"dataset/images/train/{img_name}", img)

    # Save label
    with open(f"dataset/labels/train/img_{i}.txt", "w") as f:
        f.write("\n".join(label_lines))

print("Dataset Generated Successfully with diverse shapes, realistic colors, and textures!")
```

```python
import cv2
import numpy as np
import os
import random

# Create folders
os.makedirs("dataset/images/train", exist_ok=True)
os.makedirs("dataset/labels/train", exist_ok=True)

img_size = 640
num_images = 100

for i in range(num_images):

    # Randomly select a skin background color
    background_color = random.choice(skin_background_colors)
    img = np.full((img_size, img_size, 3), background_color, dtype=np.uint8)

    num_objects = random.randint(1, 4)
    label_lines = []

    for _ in range(num_objects):
        shape_type = random.choice(['rectangle', 'circle', 'ellipse', 'polygon'])

        # Randomly select a lesion color
        base_lesion_color = random.choice(lesion_colors)
        color = base_lesion_color # Assign base_lesion_color to color for consistency

        if shape_type == 'rectangle':
            # Random rectangle (object)
            w, h = random.randint(40, 150), random.randint(40, 150)
            x = random.randint(0, img_size - w)
            y = random.randint(0, img_size - h)

            cv2.rectangle(img, (x, y), (x + w, y + h), base_lesion_color, -1)

            # Add concentric rectangles for texture
            if w > 10 and h > 10:
                for _ in range(random.randint(1, 2)): # One or two concentric shapes
                    reduce_w, reduce_h = random.randint(5, 15), random.randint(5, 15)
                    inner_w, inner_h = w - reduce_w, h - reduce_h
                    if inner_w > 0 and inner_h > 0: # Ensure dimensions are positive
                        inner_x = x + reduce_w // 2
                        inner_y = y + reduce_h // 2
                        cv2.rectangle(img, (inner_x, inner_y), (inner_x + inner_w, inner_y + inner_h), get_varied_color(base_lesion_color), -1)

            # YOLO format (normalized)
            x_center_yolo = (x + w / 2) / img_size
            y_center_yolo = (y + h / 2) / img_size
            width_yolo = w / img_size
            height_yolo = h / img_size

        elif shape_type == 'circle':
            r = random.randint(20, 75) # Radius
            cx = random.randint(r, img_size - r)
            cy = random.randint(r, img_size - r)

            cv2.circle(img, (cx, cy), r, base_lesion_color, -1)

            # Add concentric circles for texture
            if r > 5:
                for _ in range(random.randint(1, 2)): # One or two concentric shapes
                    inner_r = r - random.randint(3, 10)
                    if inner_r > 0: # Ensure radius is positive
                        cv2.circle(img, (cx, cy), inner_r, get_varied_color(base_lesion_color), -1)

            x_center_yolo = cx / img_size
            y_center_yolo = cy / img_size
            width_yolo = (2 * r) / img_size
            height_yolo = (2 * r) / img_size

        elif shape_type == 'ellipse':
            major_axis = random.randint(30, 100)
            minor_axis = random.randint(15, min(major_axis - 1, 90))
            cx = random.randint(major_axis, img_size - major_axis)
            cy = random.randint(major_axis, img_size - major_axis)
            angle = random.randint(0, 180) # Angle in degrees

            cv2.ellipse(img, (cx, cy), (major_axis, minor_axis), angle, 0, 360, base_lesion_color, -1)

            # Add concentric ellipses for texture
            if major_axis > 5 and minor_axis > 5:
                for _ in range(random.randint(1, 2)): # One or two concentric shapes
                    inner_major_axis = major_axis - random.randint(3, 10)
                    inner_minor_axis = minor_axis - random.randint(3, 10)
                    if inner_major_axis > 0 and inner_minor_axis > 0: # Ensure dimensions are positive
                        cv2.ellipse(img, (cx, cy), (inner_major_axis, inner_minor_axis), angle, 0, 360, get_varied_color(base_lesion_color), -1)

            x_center_yolo = cx / img_size
            y_center_yolo = cy / img_size
            width_yolo = (2 * max(major_axis, minor_axis)) / img_size
            height_yolo = (2 * max(major_axis, minor_axis)) / img_size

        elif shape_type == 'polygon':
            num_vertices = random.randint(3, 7)
            points = []
            # Define a rough bounding box for the polygon vertices
            poly_cx = random.randint(50, img_size - 50)
            poly_cy = random.randint(50, img_size - 50)
            poly_size = random.randint(30, 80) # Max deviation from center

            min_x, min_y = img_size, img_size
            max_x, max_y = 0, 0

            for _ in range(num_vertices):
                px = random.randint(max(0, poly_cx - poly_size), min(img_size - 1, poly_cx + poly_size))
                py = random.randint(max(0, poly_cy - poly_size), min(img_size - 1, poly_cy + poly_size))
                points.append([px, py])
                min_x = min(min_x, px)
                min_y = min(min_y, py)
                max_x = max(max_x, px)
                max_y = max(max_y, py)

            points = np.array(points, np.int32)
            points = points.reshape((-1, 1, 2)) # Reshape for cv2.fillPoly

            cv2.fillPoly(img, [points], base_lesion_color)

            # Add noise for polygon texture
            mask = np.zeros(img.shape[:2], dtype=np.uint8)
            cv2.fillPoly(mask, [points], 255)

            noise_layer = np.random.randint(-20, 21, img.shape, dtype=np.int16) # Random integers between -20 and 20
            textured_img = img.astype(np.int16) + noise_layer
            textured_img = np.clip(textured_img, 0, 255).astype(np.uint8)

            img[mask == 255] = textured_img[mask == 255]

            x_center_yolo = ((min_x + max_x) / 2) / img_size
            y_center_yolo = ((min_y + max_y) / 2) / img_size
            width_yolo = (max_x - min_x) / img_size
            height_yolo = (max_y - min_y) / img_size

        label_lines.append(f"0 {x_center_yolo} {y_center_yolo} {width_yolo} {height_yolo}")

    # Save image
    img_name = f"img_{i}.jpg"
    cv2.imwrite(f"dataset/images/train/{img_name}", img)

    # Save label
    with open(f"dataset/labels/train/img_{i}.txt", "w") as f:
        f.write("\n".join(label_lines))

print("Dataset Generated Successfully with realistic colors and textures!")
```
